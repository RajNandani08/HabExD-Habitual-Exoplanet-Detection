```python
# split datasets
x_train = train_data.drop(["LABEL"],axis = 1)
y_train = train_data["LABEL"]
x_test = test_data.drop(["LABEL"],axis = 1)
y_test = test_data["LABEL"]
```

```python
# Data Normalization
from sklearn.preprocessing import normalize
x_train = normalized = normalize(x_train)
x_train
```

```python
x_test = normalize(x_test)
x_test
```

```python
# Apply Gausian Filter
from scipy import ndimage
x_train = ndimage.filters.gaussian_filter(x_train,sigma = 10)
x_test = ndimage.filters.gaussian_filter(x_test,sigma = 10)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
x_train = StandardScaler().fit_transform(x_train)
x_test = StandardScaler().fit_transform(x_test)
```

```python
# Dimensionality Reduction by Principle Component Analysis(PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
X_Train = pca.fit_transform(x_train)
X_Test = pca.fit_transform(x_test)
```

```python
total = sum(pca.explained_variance_)
```

```python
k = 0
current_variance = 0
while current_variance/total < 0.999:
    current_variance += pca.explained_variance_[k]
    k += 1
print(k)
```

```python
# Apply PCA with n_components
pca = PCA(n_components = k)
x_train = pca.fit_transform(x_train)
x_test = pca.fit_transform(x_test)
```

```python
def model(classifier,x_train,y_train,x_test,y_test):
    # fit the model
    model = classifier.fit(x_train,y_train)
    predict = model.predict(x_test)
    # cross-validation
    accuracies = cross_val_score(estimator = classifier,X = x_train,y = y_train,cv = 5,n_jobs = -1)
    mean = accuracies.mean()
    variance = accuracies.std()
    print("*****"+str(classifier)+"******")
    print("\n")
    print("Accuracy mean : "+str(mean))
    print("Accuracy variance : "+str(variance))
    
    print("Accuracy Score : ",accuracy_score(y_test,predict))
    #Classification report
    classified_report = classification_report(y_test,predict,output_dict=True,zero_division=0)
    print("Classification Report\n",classification_report(y_test,predict,zero_division = 0))
    #Confusion Matrix
    plt.figure(figsize=(13,10))
    plt.subplot(221)
    sns.heatmap(confusion_matrix(y_test,predict),annot=True,cmap="viridis",fmt =  "d",linecolor="k",linewidths=3)
    plot.title("CONFUSION MATRIX FOR"+' '+str(classifier))
    print("ULTIMATE RESULT")
    print("No of Exoplanets : ",classified_report["1"]['support'])
    print("No of Non-Exoplanets : ",classified_report["0"]['support'])
    
    # roc-curve
    predict_prob = model.predict_proba(x_test)
    fpr1,tpr1,thresh1 = roc_curve(y_test,predict_prob[:,1],pos_label=1)
    random_probs = [0 for i in range(len(y_test))]
    p_fpr,p_tpr,_ = roc_curve(y_test,random_probs,pos_label=1)
    auc_score = roc_auc_score(y_test,predict_prob[:,1])
    
    # plot roc curves
    plt.subplot(222)
    plt.plot(fpr1,tpr1,linestyle='-',color='red',label=("Area under curve : ",auc(fpr1,tpr1)))
    plt.plot(p_fpr,p_tpr,linestyle='-',color='blue')
    plt.title('ROC curve')
    plt.xlabel('FALSE POSITIVE RATE')
    plt.ylabel('TRUE POSITIVE RATE')
    plt.legend(loc='best')
    plt.show()
    return model
import matplotlib.pyplot as plot
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score,recall_score,roc_curve,auc,f1_score,roc_auc_score,confusion_matrix,accuracy_score,classification_report
tested_model1 = model(DecisionTreeClassifier(max_depth=5,random_state=10),xv_train,yv_train,xv_test[:],yv_test[:])
tested_model2 = model(SVC(probability=True),xv_train,yv_train,xv_test[:],yv_test[:])
tested_model3 = model(LogisticRegression(),xv_train,yv_train,xv_test[:],yv_test[:])
tested_model4 = model(RandomForestClassifier(n_estimators=100,max_features="sqrt",n_jobs=-1,random_state=42,max_depth = 5),xv_train,yv_train,xv_test[:],yv_test[:])
tested_model5 = model(GradientBoostingClassifier(n_estimators=100,learning_rate=1,max_depth=5),xv_train,yv_train,xv_test[:],yv_test[:])
```
